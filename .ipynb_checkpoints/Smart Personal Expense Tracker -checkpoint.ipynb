```python
import pandas as pd
```

```python
budget_df = pd.read_csv('Budget.csv')
transaction_df = pd.read_csv('personal_transactions.csv')
```

```python
print(budget_df.head())
print(transaction_df.head())
```

```python
transaction_df['Date'] = pd.to_datetime(transaction_df['Date'], format='%m/%d/%Y')
print(transaction_df.dtypes)  # Check the types after conversion
```

```python
print(budget_df.isnull().sum())
print(transaction_df.isnull().sum())
```

```python
transaction_df.ffill(inplace=True)
```

```python
merged_df = pd.merge(transaction_df, budget_df, on='Category', how='left')
print(merged_df.head())
```

```python
merged_df['Overspent'] = merged_df['Amount'] > merged_df['Budget']
print(merged_df[['Category', 'Amount', 'Budget', 'Overspent']].head())
```

```python
import matplotlib.pyplot as plt

merged_df.groupby('Category')['Amount'].sum().plot(kind='bar', figsize=(10,6))
plt.title('Total Spending by Category')
plt.ylabel('Amount Spent')
plt.show()
```

```python
merged_df['Budget'].fillna(0, inplace=True)
```

```python
overspent = merged_df[merged_df['Overspent'] == True]
plt.figure(figsize=(10,6))
plt.bar(overspent['Category'], overspent['Amount'], label='Actual Spending')
plt.bar(overspent['Category'], overspent['Budget'], label='Budget', alpha=0.6)
plt.title('Overspending by Category')
plt.ylabel('Amount')
plt.legend()
plt.show()
```

```python
transaction_df.set_index('Date', inplace=True)
monthly_spending = transaction_df.resample('M')['Amount'].sum()
monthly_spending.plot(kind='line', figsize=(10,6))
plt.title('Spending Over Time')
plt.ylabel('Amount Spent')
plt.show()
```

```python
# Filter for overspent categories
overspent = merged_df[merged_df['Overspent'] == True]
```

```python
# Plotting Actual Spending vs. Budget
plt.figure(figsize=(10,6))
plt.bar(overspent['Category'], overspent['Amount'], label='Actual Spending', color='orange')
plt.bar(overspent['Category'], overspent['Budget'], label='Budget', alpha=0.6, color='blue')
plt.title('Overspending by Category')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
monthly_spending = transaction_df.resample('M')['Amount'].sum()
```

```python
# Plotting Spending Over Time
plt.figure(figsize=(10,6))
monthly_spending.plot(kind='line', marker='o')
plt.title('Monthly Spending Over Time')
plt.ylabel('Amount Spent')
plt.xlabel('Month')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```

```python
# Total Spending by Category
total_spending_by_category = merged_df.groupby('Category')['Amount'].sum()
```

```python
plt.figure(figsize=(10,6))
total_spending_by_category.sort_values().plot(kind='barh', color='lightgreen')
plt.title('Total Spending by Category')
plt.xlabel('Amount Spent')
plt.ylabel('Category')
plt.show()
```

```python
# Creating a pivot table for monthly spending by category
monthly_category_spending = transaction_df.pivot_table(values='Amount', 
                                                         index=transaction_df.index.month, 
                                                         columns='Category', 
                                                         aggfunc='sum')
```

```python
# Plotting Monthly Spending by Category
monthly_category_spending.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Monthly Spending by Category')
plt.ylabel('Amount Spent')
plt.xlabel('Month')
plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
transaction_df.reset_index(inplace=True)
```

```python
# Now, you can check the updated DataFrame
print(transaction_df.head())
```

```python
#1. Add a 'Month' feature to the transaction DataFrame
transaction_df['Month'] = transaction_df['Date'].dt.month
```

```python
# 2. Calculate total spending per month
monthly_spending = transaction_df.groupby('Month')['Amount'].sum().reset_index()
monthly_spending.rename(columns={'Amount': 'Total_Spending'}, inplace=True)
```

```python
# Merge the monthly spending back to the original DataFrame if needed
transaction_df = transaction_df.merge(monthly_spending, on='Month', how='left')
```

```python
# 3. Calculate percentage of total spending by category
total_spending = merged_df['Amount'].sum()
category_percentage = merged_df.groupby('Category')['Amount'].sum() / total_spending * 100
```

```python
# Create a new DataFrame to store the percentages
category_percentage_df = category_percentage.reset_index()
category_percentage_df.columns = ['Category', 'Percentage_of_Total_Spending']
```

```python
# Merge the category percentage with the merged DataFrame
merged_df = merged_df.merge(category_percentage_df, on='Category', how='left')
```

```python
# 4. (Optional) Create lag feature for previous month's spending (if you want to analyze time patterns)
transaction_df['Previous_Amount'] = transaction_df['Amount'].shift(1)
```

```python
# Check the updated DataFrame
print(merged_df.head())
```

```python
# Total spending by month
monthly_spending = transaction_df.groupby('Month')['Amount'].sum()
monthly_spending.plot(kind='bar', figsize=(10, 6))
plt.title('Total Spending by Month')
plt.ylabel('Total Amount Spent')
plt.xlabel('Month')
plt.xticks(rotation=0)
plt.show()
```

```python
# Spending trends over time
transaction_df.groupby('Date')['Amount'].sum().plot(figsize=(10, 6))
plt.title('Spending Trends Over Time')
plt.ylabel('Total Amount Spent')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.show()
```

```python
# Percentage of total spending by category
category_spending = transaction_df.groupby('Category')['Amount'].sum()
category_spending_percentage = (category_spending / category_spending.sum()) * 100
category_spending_percentage.plot(kind='pie', figsize=(10, 6), autopct='%1.1f%%')
plt.title('Percentage of Total Spending by Category')
plt.ylabel('')
plt.show()
```
